```python
def find_word (grid, word):
    rows = len(grid)
    cols = len(grid[0])
    word_len = len(word)
    directions = [
        (0, 1),   # Horizontal, left-to-right
        (0, -1),  # Horizontal, right-to-left
        (1, 0),   # Vertical, top-to-bottom
        (-1, 0),  # Vertical, bottom-to-top
        (1, 1),   # Diagonal, top-left to bottom-right
        (-1, -1), # Diagonal, bottom-right to top-left
        (-1, 1),  # Diagonal, bottom-left to top-right
        (1, -1)   # Diagonal, top-right to bottom-left
    ]

    found_place = []

    def in_bounds (x,y):
        return 0 < x < row and 0 < y < cols

    for row in range(rows): #parcours les lignes
        for col in range(cols): #parcours les colonnes
            for dx, dy in directions: #applique les directions
                found = True 
                for i in range(word_len): #regarde les 4 positions suivantes
                    nx, ny = row + i * dx, col + i * dy #enregistre la position actuel
                    if not in_bounds(nx, ny) or grid[nx][ny] != word[i]: #si la position n'est ni dans Xmas ou dans les limites de la grid, false
                        found = False
                        break
                if found:
                    found_place.append((row, col, dx, dy))
    return found_place
```

```python
def find_word_in_grid(grid, word):
    rows = len(grid)
    cols = len(grid[0])
    word_len = len(word)
    directions = [
        (0, 1),   # Horizontal, left-to-right
        (0, -1),  # Horizontal, right-to-left
        (1, 0),   # Vertical, top-to-bottom
        (-1, 0),  # Vertical, bottom-to-top
        (1, 1),   # Diagonal, top-left to bottom-right
        (-1, -1), # Diagonal, bottom-right to top-left
        (-1, 1),  # Diagonal, bottom-left to top-right
        (1, -1)   # Diagonal, top-right to bottom-left
    ]
    found_positions = []

    def in_bounds(x, y):
        return 0 <= x < rows and 0 <= y < cols

    for row in range(rows):
        for col in range(cols):
            for dx, dy in directions:
                found = True
                for i in range(word_len):
                    nx, ny = row + i * dx, col + i * dy
                    #print(nx, ny)
                    if not in_bounds(nx, ny) or grid[nx][ny] != word[i]:
                        found = False
                        break
                if found:
                    found_positions.append((row, col, dx, dy))
                    #print(found_positions)
    return found_positions

# Find all instances of 'XMAS'
positions = find_word_in_grid(grid, "XMAS")

#nombre d'iteration
len(positions)
```
